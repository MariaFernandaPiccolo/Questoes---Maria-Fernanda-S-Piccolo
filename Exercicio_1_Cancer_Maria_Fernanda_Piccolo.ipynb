Controle Estatístico de Processos (CEP)

```python
import matplotlib.pyplot as plt
dframe_plot = dframe5.drop('Categoria', axis=1)
fig, ax = plt.subplots(9, figsize=(15,10))
for i in range(dframe_plot.shape[1]):
    x = dframe_plot.iloc[:, i]
    mean = x.median() #média
    std = x.std(ddof=1) #desvio padrão
    LCL = mean - std #Limite inferior de controle
    UCL = mean + std #Limite superior de controle

    x.plot(ax=ax[i])
    ax[i].legend()
    ax[i].axhline(mean, c='k')
    ax[i].axhline(LCL, ls='--', c='r')
    ax[i].axhline(UCL, ls='--', c='r')
    ax[i].set_title(dframe_plot.columns[i])
```
