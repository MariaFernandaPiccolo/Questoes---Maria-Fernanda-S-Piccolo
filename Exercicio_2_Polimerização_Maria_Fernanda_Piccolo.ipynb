Controle Estatístico de Processos (CEP)

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(dframe.shape[1], figsize=(15,12))
for i in range(dframe.shape[1]):
    x = dframe.iloc[:, i]
    mean = x.median() #média
    std = x.std(ddof=1) #desvio padrão
    LCL = mean - std #Limite inferior de controle
    UCL = mean + std #Limite superior de controle

    x.plot(ax=ax[i])
    ax[i].legend([f'{dframe.columns[i]}'])
    ax[i].axhline(mean, color='k', linestyle='-')
    ax[i].axhline(LCL, color='r', linestyle='--')
    ax[i].axhline(UCL, color='r', linestyle='--')
    ax[i].set_title(f"Gráfico de Controle - {dframe.columns[i]}")


plt.tight_layout()
plt.show()
```
